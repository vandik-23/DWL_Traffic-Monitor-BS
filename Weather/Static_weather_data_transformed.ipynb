```python
import numpy as np
import pandas as pd
```

```python
# Hourly data for 2020
data = np.loadtxt(r"C:\Users\natal\OneDrive\Documents\GitHub\DWLadies\Stampfenbachstrasse_107-hour 2020.dat", delimiter=',')
df2020 = pd.DataFrame(data, columns=["year", "month", "day", "hour", "solarradiation", 
                                   "temperature", "relativehumidityhc", "precipitation", 
                                  "windspeedultrasonic", "winddirultrasonic"])
df2020
```

```python
# Hourly data for 2021
data = np.loadtxt(r"C:\Users\natal\OneDrive\Documents\GitHub\DWLadies\Stampfenbachstrasse_107-hour 2021.dat", delimiter=',')
df2021 = pd.DataFrame(data, columns=["year", "month", "day", "hour", "solarradiation", 
                                   "temperature", "relativehumidityhc", "precipitation", 
                                  "windspeedultrasonic", "winddirultrasonic"])
df2021
```

```python
# Hourly data for 2022
data = np.loadtxt(r"C:\Users\natal\OneDrive\Documents\GitHub\DWLadies\Stampfenbachstrasse_107-hour 2022.dat", delimiter=',')
df2022 = pd.DataFrame(data, columns=["year", "month", "day", "hour", "solarradiation", 
                                   "temperature", "relativehumidityhc", "precipitation", 
                                  "windspeedultrasonic", "winddirultrasonic"])
df2022
```

```python
# Hourly data for 2023
data = np.loadtxt(r"C:\Users\natal\OneDrive\Documents\GitHub\DWLadies\Stampfenbachstrasse_107-hour 2022 2023.dat", delimiter=',')
df2223 = pd.DataFrame(data, columns=["year", "month", "day", "hour", "solarradiation", 
                                   "temperature", "relativehumidityhc", "precipitation", 
                                  "windspeedultrasonic", "winddirultrasonic"])
df2223
```

```python
# retrieve only data from 2023 in df2023
df2023 = df2223.groupby(["year"]).get_group(2023)
df2023
```

```python
# merge datasets
df1 = pd.merge_ordered(df2020, df2021)
df2 = pd.merge_ordered(df1, df2022)
df = pd.merge_ordered(df2, df2023)
df
```

```python
data = df
```

```python
# convert to datetime
data['timestamp']= pd.to_datetime(data[['year', 'month', 'day', 'hour']], format = '%d/%m/%y %H:%M:%S.%f', exact = True)
```

```python
# remove year month day and hour
df = data.drop(columns=['year', 'month', 'day', 'hour'])
```

```python
# rearrange columns
cols = ["timestamp", "solarradiation", "temperature", 
        "relativehumidityhc", "precipitation", 
        "windspeedultrasonic", "winddirultrasonic"]
df = df[cols]
df
```

```python
# Round timestamp
df['timestamp'] = df['timestamp'].apply(lambda x: x.round(freq='H'))
```
